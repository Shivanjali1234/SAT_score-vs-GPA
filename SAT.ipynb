```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"C:\Users\nirma\Downloads\1.01. Simple linear regression.csv")
```

```python
df
```

```python
lr=LinearRegression()
x=df.iloc[:,0].values.reshape(-1,1)
y=df.iloc[:,1].values.reshape(-1,1)
```

```python
lr.fit(x,y)
```

```python
y_predict=lr.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict,color='r')
```

```python
plt.xlabel("SAT")
plt.ylabel("GPA")
plt.title("SAT,GPA Prediction Model")
```
